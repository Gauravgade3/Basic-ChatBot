```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
from tkinter import *
import time


bot=ChatBot('mybot')


#training chatbot with conversation
conversation = [
    "Hello",
    "Hi there!",
    "How are you?",
    "I'm doing great.",
    "That is good to hear",
    "Thank you.",
    "You're welcome."
    ,'what is your name?',
    'My name is Bot.',
    'who created you?',
    'I am created by Gaurav.',
    'where are you working now?',
    'I am working at Accenture.',
    'In which institute are you studying now?',
    'I am persuing Data Science at AlmaBetter.'
    ]

trainers=ListTrainer(bot)
trainers.train(conversation)


#setting frames
main = Tk()
main.geometry("500x730")              #set size of window
main.title("My ChatBot")

#set title of chat bot
img = PhotoImage(file='robot.png')
photoL = Label(main, image=img)
photoL.pack(pady=5)
###########################################################

#show current time and date
def start():

    text=time.strftime("%d-%m-%Y   %H:%M:%S  %p")
    label.config(text=text)
    label.after(200,start)
label=Label(main,bg="#ffffff",fg="red",font=("verdana",20,"bold"))
label.pack()
start()

############################################################
#define a function to chat with bot
def ask_from_bot ():
    query=text.get()
    answer_from_bot=bot.get_response(query)
    msgs.insert(END,'you :' + query)
    
    msgs.insert(END, 'bot :' + str(answer_from_bot))
    text.delete(0,END)
    msgs.yview(END)

# create a menu bar and quit button
menubar = Menu(main)
menubar.add_command(label="Quit!", command=main.quit)
main.config(menu=menubar)
frame = Frame(main)

#set a scrollbar
sc = Scrollbar(frame)                                   #set scrollbar
msgs = Listbox(frame, width=80, height=10,yscrollcommand=sc.set)              #provide listbox
sc.pack(side=RIGHT, fill=Y)
msgs.pack(side=LEFT, fill=BOTH, pady=10)

frame.pack()
text = Entry(main, font=("Verdana", 20))  # format text
text.pack(fill=X, pady=10)

#button- ask from bot
btn = Button(main, text="Ask from bot",bg='#000000' , font=('Verdana', 20), fg='red', command = ask_from_bot)
btn.pack(expand=True,fill="both")

#binding keyboard key-(enter) with btn
def enter_function(event):
    btn.invoke()
main.bind('<Return>',enter_function)





main.mainloop()
```
